Step 1: Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2: Load the Iris dataset

```python
# Load the iris dataset
df = pd.read_csv("iris.csv")
```

Step 3: Inspect the Dataset

```python
# Print shape
print("Shape of the dataset:", df.shape)

# Print column names
print("Names of the columns:", df.columns.tolist())

# Print first 5 rows
print(df.head())
```

Step 4: Summary Statistics

```python
print("\nDataset Info:")
print(df.info())
```

```python
# Description of iris dataset
print("Summary Statistics:")
print(df.describe())
```

Step 5: Visualizations

A. Scatter Plot-- Relationship between features

```python
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

B. Histograms--Value Distributions

```python
df.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle('Histograms of Iris Features')
plt.tight_layout()
plt.show()
```

C. Box Plots--Identify Outliers

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Box Plots of Iris Features')
plt.xticks(rotation=45)
plt.show()
```
